```
# create dbase & table schema

import sqlite3
connection = sqlite3.connect("pystudents.db")
cur = connection.cursor()

sql = """
      CREATE TABLE StudentsTable (
            studentID TEXT,
            yearJoined INTEGER,
            Title TEXT,
            first_Name TEXT,
            last_Name TEXT,
            class TEXT,
            phone_Number TEXT,
            Email_addre TEXT,
            City TEXT,
            state TEXT, 
            Country TEXT,
            primary key(studentID)
            )"""
cur.execute(sql)
print("Python students Table has been created.")

connection.commit() # save to the Database
connection.close()
```

```
import sqlite3, csv 

connection = sqlite3.connect("pystudents.db") 
cursor = connection.cursor()

with open('records2.csv', 'r') as file:
      no_records = 0
      for row in file:
          cursor.execute("INSERT INTO StudentsTable VALUES (?,?,?,?,?,?,?,?,?,?,?)", row.split(","))
          connection.commit()
          no_records += 1
connection.close()
print('\n{} Records Transfered'.format(no_records))
```
